```python
%matplotlib inline

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
subsc = pd.read_csv('data/subscribers.txt', encoding = "ISO-8859-1", sep='\t')
```

```python
subsc.head()
```

```python
subsc["start_date"] = pd.to_datetime(subsc["start_date"])
```

```python
subsc["year"] = subsc["start_date"].dt.year
```

```python
subsc.head()
```

```python
subsc = subsc[(subsc["year"] == 2004) | (subsc["year"] == 2005) | (subsc["year"] == 2006)]
```

```python
subsc.tail()
```

```python
pogrupowane = subsc.groupby([subsc.year, subsc.start_date.dt.month, subsc.start_date.dt.day]).count()
```

```python

fig, ax = plt.subplots(figsize=(15,7))
pogrupowane["start_date"].unstack()
```
